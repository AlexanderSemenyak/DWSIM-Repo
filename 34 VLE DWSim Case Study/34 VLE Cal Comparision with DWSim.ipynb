Vapour/liquid equilibrium is a critical concept in chemical engineering and plays an important role in many industrial processes. It describes the equilibrium state that exists between a liquid and its vapor and determines the composition of the liquid and vapor in a system. The VLE can be calculated using several methods, including phase equilibrium diagrams, equations of state, and computer simulation methods. Understanding VLE is essential for the design and optimization of many industrial processes and is an important area of study in chemical engineering.

```python
import numpy as np
import matplotlib.pyplot as plt
Xl = np.linspace(0,1,100)
X2 = 1-Xl
```

```python
# Defining antoine constants
ANT_Acetonitrile = [14.8950,3413.10,-22.627]
ANT_Nitromethane = [14.7513,3331.70,-45.550]
```

```python
def Psat1(T):
    return np.exp(ANT_Acetonitrile[0]-ANT_Acetonitrile[1]/(T+ANT_Acetonitrile[2]))
Psat1(348.15)
```

```python
def Psat2(T):
    return np.exp(ANT_Nitromethane[0]-ANT_Nitromethane[1]/(T+ANT_Nitromethane[2]))
Psat2(348.15)
```

```python
Total_Pressure = Xl*Psat1(348.15) + X2*Psat2(348.15)
y1 = Xl*Psat1(348.15)/Total_Pressure
```

```python
# For Dew Points
Total_Pressure_Dew = 1/(Xl/Psat1(348.15)+X2/Psat2(348.15))
```

```python
x1 = Xl*Total_Pressure_Dew/Psat1(348.15)
```

```python
plt.plot(Xl,Total_Pressure)
plt.plot(Xl,Total_Pressure_Dew)
plt.xlabel('x1,y1',fontsize = 15)
plt.ylabel('P/kPa',fontsize = 15)
plt.title('Manual Calculation Plot P vs xy')
# plt.legend()
plt.grid()
plt.show()
```
